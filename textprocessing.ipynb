## Word Segmentation

```python
import nltk
from nltk.tokenize import word_tokenize

text = "The sun was setting over the horizon, casting a warm orange and pink glow over the sky. The trees rustled in the gentle breeze, their leaves rustling and swaying. Birds sang in the distance, their sweet melodies filling the air. The grass was a lush green, soft and inviting underfoot. The air was filled with the sweet scent of blooming flowers and freshly cut grass. The world seemed at peace, and all was right with the world. The people of the village went about their day, tending to their farms, chatting with their neighbors, and enjoying the simple pleasures of life. The children laughed and played, their laughter and shouts of joy filling the air. The older folk sat on their porches, watching the world go by, their faces etched with wisdom and experience. Life was good, and all was right with the world."
tokens = word_tokenize(text)
tokens
```

## Sentence Segmentation

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)
sentences
```

## Convert to lowercase

```python
text_lower = text.lower()
print(text_lower)
```

## Stop words removal

```python
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
tokens = word_tokenize(text_lower)
new_sentence_tokenized = [x for x in tokens if x not in stop_words]

# After removing the stop words
new_sentence_tokenized
```

## Stemming

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in new_sentence_tokenized]
stemmed_words
```

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in new_sentence_tokenized]
lemmatized_words
```

## Part of Speed Tagger

```python
from nltk import pos_tag

tagged_words = pos_tag(new_sentence_tokenized)
tagged_words
```
